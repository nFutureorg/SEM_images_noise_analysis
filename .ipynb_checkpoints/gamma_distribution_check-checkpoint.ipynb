<b>alpha is the shape parameter, where lower values increase the skew (a shape of 1 gives the exponential distribution, and higher numbers tend towards a normal distribution), and beta is the rate parameter, where higher values lower the mean. (see Characterization using shape α and rate β here.)</b>

<b>The mean of a gamma distribution is alpha/beta, and the standard deviation is sqrt(alpha/beta^2). To get your target mean and standard deviations, you would need shape parameters of c(400, 3402) and rate parameters of c(20, 48.61). These shape parameters would lead to distributions that would be impossible to distinguish from normal distributions at reasonable sample sizes.</b>

<b>To get positively skewed distributions, we need considerably lower shape parameters. </b>

<b>shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)</b>

<b>noise = np.random.gamma(shape, scale, (1024,786))</b>

```python
img = cv2.imread('SEM/L7_0a800855e3b88fd72a83fe7dd8257f88.jpg')
crop_img_clean = img[y:y+h, x:x+w] #Crop images
                  
# resize image
resized = cv2.resize(crop_img_clean, dim, interpolation = cv2.INTER_CUBIC)
    
rng = np.random.default_rng()
# gamma(shape,scale,size)
gauss = rng.gamma(1.,70, resized.shape)
                
resized_noisy_img = resized + gauss


    
cv2.imwrite('outsem/pollen.spd8.0098.jpg', resized_noisy_img)
```
